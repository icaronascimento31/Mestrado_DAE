```python
import pickle
import numpy as np
import pandas as pd

# =========================================================
# Carregar dados
# =========================================================
with open("loss_curves_full_energy.pkl", "rb") as f:
    data = pickle.load(f)

train_curves = data["train_curves"]
val_curves   = data["val_curves"]

# =========================================================
# Calcular médias por fold
# =========================================================
results = []

for i in range(len(train_curves)):
    train_mean = np.mean(train_curves[i])

    if val_curves[i] is not None:
        val_mean = np.mean(val_curves[i])
    else:
        val_mean = np.nan

    results.append([i + 1, train_mean, val_mean])

# =========================================================
# Criar tabela
# =========================================================
df = pd.DataFrame(results, columns=[
    "Fold",
    "Loss Médio Treino",
    "Loss Médio Validação"
])

# Média global
global_train = df["Loss Médio Treino"].mean()
global_val   = df["Loss Médio Validação"].mean()

# Linha final
df.loc[len(df)] = ["Média Geral", global_train, global_val]

# Exibir
print(df)

# Salvar CSV
df.to_csv("tabela_loss_folds.csv", index=False)
```
